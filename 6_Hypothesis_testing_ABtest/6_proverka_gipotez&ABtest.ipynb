### Применение фреймворка ICE для приоритизации гипотез. 

Проведем приоритезацию гипотез с помощью фреймворка `ICE` для этого - перемножим `impact`(силу влияния на пользователей) и `confidence`(уверенность в том что изменение повлияет на пользователей именно так) и разделим полученное произведение на `efforts`(сложность тестирования гипотезы)

```python
hypothesis['ice'] = round(
    hypothesis['impact']*hypothesis['confidence']/hypothesis['efforts'], 
    2
)

hypothesis[['hypothesis', 'ice']].sort_values(by='ice', ascending = False)
```

Приоретизация с помощью фреймворка `ICE` показала (индексы гипотез будем принимать за номер гипотезы):

- На первое место по приоритизации методом `ICE` вышла гипотеза 8 "Запустить акцию, дающую скидку на товар в день рождения". Это индивидуальный подход дающий покупателю в его праздничный день почувствовать себя особенным.

- На втором и третьем месте гипотезы 0 (о добавлении двух новых каналов трафика) и 7 (о добавлении формы подписки на все основные страницы). 

- Самый низкий приоретет у гипотез 4 (Об изменении фона главной страницы) и 3 (об изменении структуры категорий). 

Посмотрим на приоритизацию гипотез в разрезе фреймворка `RICE`.

### Применение фреймворка RICE для приоритизации гипотез. 

Проведем приоритезацию гипотез с помощью фреймворка `RICE` для этого - перемножим параметры `reach` (оценка скольких пользователей затронет изменение), `impact`(силу влияния на пользователей) и `confidence`(уверенность в том что изменение повлияет на пользователей именно так) полученное произведение разделим  на `efforts`(сложность тестирования гипотезы).

```python
hypothesis['rice'] = (
    hypothesis['reach']*hypothesis['impact']*hypothesis['confidence']
)/hypothesis['efforts']

hypothesis[['hypothesis', 'rice']].sort_values(by='rice', ascending = False)
```

### Построим график кумулятивной выручки по группам. 

Чтобы построить графики, нужно собрать кумулятивные данные.

Объявим датафрейм `cumulative_dt` со столбцами: 
* `date` — дата;
* `group` — группа A/B-теста (`A` или `B`);
* `orders` — кумулятивное количество заказов на указанную дату в указанной группе;
* `buyers` — кумулятивное количество пользователей, совершивших хотя бы один заказ, на указанную дату в указанной группе;
* `revenue` — кумулятивная выручка на указанную дату в указанной группе (средний чек);
* `visitors` — кумулятивное количество посетителей интернет-магазина на указанную дату в определённой группе.

Создадим массив уникальных пар значений дат и групп теста методом `drop_duplicates()`: 

```python
# создаем массив уникальных пар значений дат и групп теста
dates_groups = orders[['date','group']].drop_duplicates() 
```

Получим строки таблицы `orders`, дата которых меньше или равна дате элемента из `dates_groups`, а группа теста равна группе из `dates_groups`


Агрегируем значения. Вычислим максимальную дату. Для группы тоже рассчитаем максимум. Найдём число уникальных `ID` заказов и пользователей. Подсчитаем сумму средних чеков. 

Применим методы к каждой строке датафрейма: `apply(axis=1)` и отсортируем результаты по столбцам `date` и `group`.

```python
orders_agg = dates_groups.apply(
    lambda x: orders[np.logical_and(
        orders['date'] <= x['date'], 
        orders['group'] == x['group']
    )].agg({
        'date' : 'max', 
        'group' : 'max', 
        'transaction_id' : 'nunique', 
        'visitor_id' : 'nunique', 
        'revenue' : 'sum'
    }), axis=1).sort_values(by=['date','group']) 
```

Аналогично получим агрегированные кумулятивные по дням данные о посетителях интернет-магазина: 

```python
visitors_agg = dates_groups.apply(
    lambda x: visitors[np.logical_and(
        visitors['date'] <= x['date'], 
        visitors['group'] == x['group']
    )].agg({
        'date' : 'max', 
        'group' : 'max', 
        'visitors' : 'sum'
    }), axis=1).sort_values(by=['date','group']) 
```

Объединим обе таблицы в одну:

```python
# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulative_dt = orders_agg.merge(
    visitors_agg, 
    left_on=['date', 'group'], 
    right_on=['date', 'group']
)
cumulative_dt.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

print(cumulative_dt.head(5)) 
```

```python
#Проверим совпадают ли минимальная и максимальная даты в cumulative_dt 
#с первоначальными данными
min_cumulative_dt_date = cumulative_dt['date'].min()
print('Минимальная дата заказа "cumulative_dt":',min_cumulative_dt_date)

max_cumulative_dt_date = orders['date'].max()
print('Максимальная дата заказа "cumulative_dt":',max_cumulative_dt_date)
```

Данные даты в `cumulative_dt` совпадают с первоначально полученными данными.

__Построим графики кумулятивной выручки по дням и группам A/B-тестирования:__

```python
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cum_rev_a = cumulative_dt[cumulative_dt['group']=='A'][['date','revenue', 'orders']]

# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cum_rev_b = cumulative_dt[cumulative_dt['group']=='B'][['date','revenue', 'orders']]

#зададим размер графика
plt.figure(figsize=(10,6)) 

# Строим график выручки группы А
plt.plot(cum_rev_a['date'], cum_rev_a['revenue'], label='A')

# Строим график выручки группы B
plt.plot(cum_rev_b['date'], cum_rev_b['revenue'], label='B')

# Зададим параметры оформления графика
plt.grid()
plt.title('График кумулятивной выручки по дням и группам A/B-тестирования', fontsize=13)
plt.ylabel('Выручка')
plt.xlabel('Дата')
plt.legend(['Группа А', 'Группа B']) 
plt.show();
```

### Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным.

Статистическую значимость различий будем считать с помощью теста Манна-Уитни, для начала подготовим данные:

```python
# таблицы с данными по заказам на каждого пользователя
orders_by_users_a = orders[orders['group']=='A'].groupby(
    'visitor_id'
)['transaction_id'].nunique().reset_index()
orders_by_users_a.columns=['visitors_id','orders']

orders_by_users_b = orders[orders['group']=='B'].groupby(
    'visitor_id'
)['transaction_id'].nunique().reset_index()
orders_by_users_b.columns=['visitors_id','orders'] 

display(orders_by_users_a.head()) 
display(orders_by_users_b.head())  
```

Объявим переменные `sample_a` и `sample_b`, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. 

```python
sample_a = pd.concat([
    orders_by_users_a['orders'], 
    pd.Series(
        0, 
        index=np.arange(visitors[visitors['group']=='A']['visitors'].sum() - len(
            orders_by_users_a['orders']
        )), 
        name='orders')
],axis=0)

sample_b = pd.concat([
    orders_by_users_b['orders'],
    pd.Series(
        0, 
        index=np.arange( visitors[visitors['group']=='B']['visitors'].sum() - len(
            orders_by_users_b['orders']
        )), 
        name='orders')
],axis=0) 
```

__Cформируем нулевую и альетрнативную гипотезы:__

- Нулевая гипотеза: Различий в конверсии между группами нет
- Альтернативная гипотеза: Различия в конверсиях между группами есть

Критический уровень статистической значимости: 0.05

```python
#задаим уровень статистической значимости
alpha = 0.05

results = stats.mannwhitneyu(sample_a, sample_b)

#значение p_value
print('Значение p-value: {:.3f}'.format(results.pvalue))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: "Различий в конверсии между группами нет"')
else:
    print('Не получилось отвергнуть нулевую гипотезу: "Различий в конверсии между группами нет"') 
    
print('Отношение конверсии В к А: {:.1%}'.format(sample_b.mean()/sample_a.mean()-1))   
```

- `P-value` меньше заданного значения `alpha` - отвергаем нулевую гипотезув пользу альтернативной гипотезы - "Различия в конверсиях между группами есть". 

- Относительный прирост конверсии группы `В` к `А` в ~16%. Далее нужно посмотреть на результаты без аномалий.

### Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным.

__Сформулируем гипотезы:__

- Нулевая гипотеза: Рразличий в среднем чеке между группами нет.
- Альтернативная гипотеза: Различия в среднем чеке между группами есть. 

Уровень статистической значимости 0.05

```python
#задаим уровень статистической значимости
alpha = 0.05

results_2 = stats.mannwhitneyu(
     orders_a['revenue'], 
     orders_b['revenue']
)

#значение p_value
print('Значение p_value: {:.3f}'.format(results_2.pvalue))

if results_2.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: "Различий в среднем чеке между группами нет"')
else:
    print('Не получилось отвергнуть нулевую гипотезу: "Различий в среднем чеке между группами нет"') 
    
print('Относительный прирост в среднем чеке группы B к А: {:.1%}'.format(
    orders_b['revenue'].mean()/orders_a['revenue'].mean()-1)
     )  
```

- `P-value` значительно больше 0.05. Значит, причин отвергать нулевую гипотезу и считать, что в среднем чеке есть различия, нет. 
- Хотя если смотреть на относительный прирост среднего чека группы `B`, он на 28,7%  выше среднего чека группы `A`, но в этом случае могут на данные влиять аномалии.

Нужносравнить эти показатели по данным без аномалий.

### Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным.

Ранее было выявлено что аномальными значениями среди количества заказов являются более 1 заказа, а по сумме заказа выбрана граница аномальных значений свыше 54 тыс. (округленное 53 904) усл.ед., заказы превышающие это колличество и сумму составляют не более 1%, их смело можно удалить, и повторно проверить гипотезы о разности групп А/В-теста.

Подготовим данные к проверке гипотез, убрав аномалии.

```python
# найдем ID пользователей с аномальными заказами (по стоимости и количеству)
users_many_orders = pd.concat(
    [
        orders_by_users_a[orders_by_users_a['orders'] >= np.percentile(
            orders_by_users_a['orders'], 
            99
        )]['visitors_id'],
        orders_by_users_b[orders_by_users_b['orders'] >= np.percentile(
            orders_by_users_b['orders'], 
            99
        )]['visitors_id'],
    ],
    axis=0,
)

users_expens_orders = orders[orders['revenue'] > np.percentile(
    orders['revenue'], 99
)]['visitor_id']

anorm_users = (
    pd.concat([users_many_orders, users_expens_orders], axis=0)
    .drop_duplicates()
    .sort_values()
)

#отфильтруем данные в группе А 
sample_filt_a = pd.concat(
    [
        orders_by_users_a[
            np.logical_not(orders_by_users_a['visitors_id'].isin(anorm_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors[visitors['group']=='A']['visitors'].sum() - len(
                    orders_by_users_a['orders']
                )),
            name='orders',
        ),
    ],
    axis=0,
)

#отфильтруем данные в группе В
sample_filt_b = pd.concat(
    [
        orders_by_users_b[
            np.logical_not(orders_by_users_b['visitors_id'].isin(anorm_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors[visitors['group']=='B']['visitors'].sum() - len(
                    orders_by_users_b['orders']
                )
            ),
            name='orders',
        ),
    ],
    axis=0,
)  
```

Применим статистический критерий Манна-Уитни к полученным выборкам и проверим гипотезы:

__Обозначим гипотезы:__

- Нулевая гипотеза: Различий в конверсии между группами нет
- Альтернативная гипотеза: Различия в конверсиях между группами есть

Критический уровень статистической значимости: 0.05

```python
#задаим уровень статистической значимости
alpha = 0.05

result_3 = stats.mannwhitneyu(sample_filt_a, sample_filt_b)[1]

#значение p_value
print('Значение p-value: {:.3f}'.format(result_3)) 

if result_3 < alpha:
    print('Отвергаем нулевую гипотезу: "Cтатистически значимых различий в конверсии между группами нет.')
else:
     print('Не получилось отвергнуть нулевую гипотезу: "Различия в конверсиях между группами есть"')  
    
print('Отношение конверсии В к А, по очищенным данным: {:.1%}'.format(sample_filt_b.mean()/sample_filt_a.mean()-1))
```

- Значение `p-value`: 0.028, меньше статистической значимости. 
- Отвергаем гипотезу о равенстве конверсий.
- Относительный прирост конверсии `В` к `А` около 12%.

Значения после фильтрации незначительно изменились, `p-value` снизился, и разница между сегментами уменьшилась с 16% до 11.9%. Конверсия у группы `В` лучше чем у группы `А`.

### Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным.

__Сформулируем гипотезы:__

- Нулевая гипотеза: Рразличий в среднем чеке между группами нет.
- Альтернативная гипотеза: Различия в среднем чеке между группами есть. 

Уровень статистической значимости 0.05

```python
#задаим уровень статистической значимости
alpha = 0.05

results_4 = stats.mannwhitneyu(
    orders_a[np.logical_not(orders_a['visitor_id'].isin(anorm_users))]['revenue'],
    orders_b[np.logical_not(orders_b['visitor_id'].isin(anorm_users))]['revenue']
)

#значение p_value
print('Значение p_value: {:.3f}'.format(results_4.pvalue))

if results_4.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: "Различий в среднем чеке между группами нет"')
else:
    print('Не получилось отвергнуть нулевую гипотезу: "Различий в среднем чеке между группами нет"') 
    
print(
    'Относительный прирост в среднем чеке группы B к А , по очищенным данным: {:.1%}'
    .format(orders_b[np.logical_not(
        orders_b['visitor_id'].isin(anorm_users)
    )]['revenue'].mean()/ orders_a[np.logical_not(
        orders_a['visitor_id'].isin(anorm_users)
    )]['revenue'].mean() - 1)
) 
```
